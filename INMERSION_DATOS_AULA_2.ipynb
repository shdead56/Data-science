2. Precio del m2 por barrio y hacer el gráfico más adecuado para esta nueva variable.

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
seleccionado = inmuebles[inmuebles['Tipo'].isin(['Apartamento', 'Local', 'Casa'])]
grafica = sns.histplot(data=seleccionado, x='Precio_Millon', kde=True, hue='Tipo')
grafica.set_title('Distribución de Valores de los inmuebles en Bogotá',fontsize=20, fontweight='bold')
plt.ylabel('CONTEO DE VALORES',fontsize=15)
plt.xlabel('PRECIO EN MILLONES',fontsize=15)
plt.xlim((50,1000))
plt.show()
```

```python
inmuebles['Precio'] = inmuebles['Precio'].astype(int)
inmuebles['m2'] = inmuebles.Precio/inmuebles.Area
inmuebles.sample(3)
```

```python
prom = inmuebles.groupby('Tipo')['m2'].mean()
import matplotlib.pyplot as plt

# crear el gráfico de barras
plt.bar(prom.index, prom.values)

# agregar títulos y etiquetas de ejes
plt.title('Promedio de Precios para cada Tipo de Inmueble')
plt.xlabel('Tipo de Inmueble')
plt.ylabel('Precio Promedio')

# mostrar el gráfico
plt.show()
```
